```python
df1 = df[df['City'] == 'Yangon']
df2 = df[df['City'] == 'Mandalay']
df3 = df[df['City'] == 'Naypyitaw']

df1[df1['Branch'] != 'A']
df2[df2['Branch'] != 'B']
df3[df3['Branch'] != 'C']
```

```python
time = pd.to_datetime(df["Time"])
df["Time(H)"] = time.dt.hour
```

```python
date = pd.to_datetime(df["Date"])
df["Day"] = date.dt.day
df["Month"] = date.dt.month
```

```python
df["gross income"] = df["gross income"].round().astype(int)
```

```python
df.head()
```

```python
df.drop(
    columns=[
        "Invoice ID",
        "City",
        "gross margin percentage",
        "Tax 5%",
        "Unit price",
        "Date",
        "Quantity",
        "Total",
        "Time",
        "cogs",
    ],
    inplace=True,
)
```

```python
df = df[
    [
        "Branch",
        "Customer type",
        "Gender",
        "Product line",
        "Payment",
        "Time(H)",
        "Day",
        "Month",
        "gross income",
        "Rating",
    ]
]
```

```python
df.head()
```

```python
numerical = ["Day", "gross income", "Rating"]

fig, ax = plt.subplots(len(numerical), 1, figsize=(8, 12))

palette = sns.palettes.color_palette("bright", 6)

for i, col in enumerate(numerical):
    sns.histplot(data=df, x=col, ax=ax[i], kde=True, color=palette[i + 1])
    ax[i].set_title(f'Histogram of {col}')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(len(numerical), 1, figsize=(8, 10))

palette = sns.palettes.color_palette("bright", 6)

for i, col in enumerate(numerical):
    sns.boxplot(data=df, x=col, ax=ax[i], color=palette[i + 1])
    ax[i].set_title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, hue="Branch", x="Customer type", width=0.5, palette="coolwarm")
plt.legend(loc="lower right")
plt.show()
```

```python
ax = sns.countplot(data=df, x="Customer type", hue="Gender", palette="coolwarm")
for p in ax.patches:
    count = p.get_height()
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(f"{count}", xy=(x, y), ha="center", va="bottom")


plt.show()
```

```python
df["Product line"].value_counts().to_frame().rename(columns={"Product line": "count"})
```

```python
ax = sns.countplot(
    data=df,
    y="Product line",
    palette="coolwarm",
    order=df["Product line"].value_counts().index,
    hue="Gender",
)
for p in ax.patches:
    count = p.get_width()
    ax.annotate(
        f"{int(count)}", xy=(count + 1, p.get_y() + p.get_height() / 2), va="center"
    )
plt.show()
```

```python
g = sns.FacetGrid(df, col="Gender", aspect=2, height=2, sharey=True, legend_out=True)
g.map(sns.countplot, "Product line", palette="coolwarm")
for ax in g.axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

plt.show()
```

```python
plt.pie(
    df["Payment"].value_counts(),
    labels=df["Payment"].unique(),
    autopct="%.2f",
    colors=["#E2BFD9", "#C8A1E0", "#674188"],
)
plt.show()
```

```python
ax = sns.countplot(data=df, x="Payment", palette="coolwarm", hue="Gender")
for p in ax.patches:
    ax.annotate(
        f"{p.get_height()}",
        (p.get_x() + p.get_width() / 2, p.get_height() + 1),
        ha="center",
    )
plt.legend(loc="lower left")
```

```python
order = (
    df.groupby("Product line")["gross income"].sum().sort_values(ascending=False).index
)
ax = sns.barplot(
    data=df,
    x="gross income",
    y="Product line",
    palette="coolwarm",
    estimator=sum,
    ci=None,
    order=order,
)
for container in ax.containers:
    ax.bar_label(container)
```

```python
ratings_by_product_line = df.groupby('Product line')['Rating'].mean().reset_index().sort_values(by='Rating', ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Rating', y='Product line', data=ratings_by_product_line,palette='coolwarm')
plt.title('Average Customer Ratings by Product Line')
for index, value in enumerate(ratings_by_product_line['Rating']):
    plt.text(value, index, str(round(value, 2)), va='center')
```
